# Data.gov.sg dataset

GES data is available at [data.gov.sg](https://beta.data.gov.sg/datasets/d_3c55210de27fcccda2ed0c63fdd2b352/view). However it is only up till 2021. 

File is downloaded and placed in `../data/GES_raw_dataGov_2021.csv`. Using this instead of the spreadsheet version because the spreadsheet version only has *FT Employment Rate*, *Mean Gross Income* and *Median Gross Income*. Missing these features:
- Overall Employment Rate
- Basic Monthly Salary Fields (Mean & Median)
- 25th Percentile Gross Monthly Salary
- 75th Percentile Gross Monthly Salary

However the spreadsheet version is updated till 2023. 

Will need to manually append the relevant data from external sources to the current data.gov.sg table which only has information up till 2021.
